```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('shift_data.csv')
```

```python
df.head()
```

```python
df['ID Number'].fillna(method='ffill', inplace=True)
```

```python
df=df[pd.notnull(df['In'])]
```

```python
df.head()
```

```python
df['Punch1']=df.Date + " " + df.In
df['Punch1'] = pd.to_datetime(df['Punch1'], format='%m/%d/%Y %H:%M')

df['Punch2']=df.Date + " " + df.Out
df['Punch2'] = pd.to_datetime(df['Punch2'], format='%m/%d/%Y %H:%M')

df['Punch3']=df.Date + " " + df['In.1']
df['Punch3'] = pd.to_datetime(df['Punch3'], format='%m/%d/%Y %H:%M')

df['Punch4']=df.Date + " " + df['Out.1']
df['Punch4'] = pd.to_datetime(df['Punch4'], format='%m/%d/%Y %H:%M')
```

```python
df.head()
```

```python
df['Punch3a'] = df.groupby(['ID Number','Date']).Punch1.shift(-1)
df['Punch4a'] = df.groupby(['ID Number','Date']).Punch2.shift(-1)
```

```python
df.head()
```

```python
df['Punch1Adj'] = df.Punch1

df['Punch2Adj'] = df.Punch2
df.loc[df.Punch2 < df.Punch1, 'Punch2Adj'] = df.Punch2 + pd.Timedelta(days=1)
df.Punch2Adj = pd.to_datetime(df['Punch2Adj'], format='%m/%d/%Y %H:%M')

df['Punch3Adj'] = df.Punch3a
df.loc[df.Punch3a < df.Punch1, 'Punch3Adj'] = df.Punch3a + pd.Timedelta(days=1)
df.Punch3Adj = pd.to_datetime(df['Punch3Adj'], format='%m/%d/%Y %H:%M')

df['Punch4Adj'] = df.Punch4a
df.loc[df.Punch4a < df.Punch1, 'Punch4Adj'] = df.Punch4a + pd.Timedelta(days=1)
df.Punch4Adj = pd.to_datetime(df['Punch4Adj'], format='%m/%d/%Y %H:%M')
```

```python
df.head()
```

```python
#Sort the data by employee, data, and time
df=df.sort_values(by=["ID Number","Date","Punch1"])
```

```python
#Renumber the index to reflect the new order
df = df.reset_index(drop=True)
```

```python
#Delete rows where the data from Punch 1 and 2 was moved to Punch 3 and 4
df.drop(df[df.Punch1 == df['Punch3a'].shift()].index, inplace=True)
```

```python
df.head()
```

```python
#Calculate the length of time worked before lunch
df['Shift_PreLunch']= df.Punch2Adj - df.Punch1Adj
```

```python
#Calculate the length of each lunch break
df['LunchBreak'] = 0
df.loc[df['Punch3Adj'].notnull(),'LunchBreak']=df.Punch3Adj - df.Punch2Adj
df.LunchBreak = pd.to_timedelta(df['LunchBreak'])
```

```python
#Calculate the length of time worked after lunch
df['Shift_PostLunch'] = 0
df.loc[df['Punch4Adj'].notnull(),'Shift_PostLunch']=df.Punch4Adj - df.Punch3Adj
df.Shift_PostLunch = pd.to_timedelta(df['Shift_PostLunch'])
```

```python
#Calculate the length of the entire shift
df['ShiftTotal'] = df.Shift_PreLunch
df.loc[df['Shift_PostLunch'].notnull(),'ShiftTotal']=df.Shift_PreLunch + df.Shift_PostLunch
df.ShiftTotal = pd.to_timedelta(df['ShiftTotal'])
```

```python
#Determine whether an employee was entitled to lunch (shift >= 5 hours), but did not take one
df['MissingLunch1'] = ''
df.loc[(df['ShiftTotal']>='0 days 05:00:00') &
       (df['LunchBreak']=='0 days 00:00:00'),       
       'MissingLunch1'] = 'missing'
```

```python
#Determine whether an employee was entitled to a second lunch (shift >= 11 hours), but did not take one
df['MissingLunch2'] = ''
df.loc[(df['ShiftTotal']>='0 days 11:00:00'),'MissingLunch2'] = 'missing'
```

```python
#Identify lunch breaks of exactly 30 minutes in length
df['PerfectLunch'] = ''
df.loc[df['LunchBreak']=='0 days 00:30:00','PerfectLunch'] = '30Minute'
```

```python
#Determine whether a lunch break was taken early (< 2 hrs from shift start)
df['EarlyLunch'] = ''
df.loc[(df['LunchBreak']!='0 days 00:00:00') &
       (df['Shift_PreLunch']>'0 days 00:00:00') &
       (df['Shift_PreLunch']<'0 days 02:00:00'),
       'EarlyLunch'] = 'early'
```

```python
#Determine whether a lunch break was taken late (>5 hrs after shift start)
df['LateLunch'] = ''
df.loc[(df['LunchBreak']!='0 days 00:00:00') &
    (df['Shift_PreLunch']>'0 days 05:00:00'),
    'LateLunch'] = 'late'
```

```python
#Create new df with only the columns I'm interested in
df2=df[['ID Number','Punch1Adj','Punch2Adj','Punch3Adj','Punch4Adj','Shift_PreLunch',
                     'LunchBreak','Shift_PostLunch','ShiftTotal','PerfectLunch','MissingLunch1','MissingLunch2',
                     'EarlyLunch','LateLunch']].copy()
```

```python
df2.head()
```

```python
df2.to_csv('shift_output.csv', encoding='utf-8')
```
